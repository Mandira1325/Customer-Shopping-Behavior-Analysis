```python
#now here review rating has 37 null values so we need to update that value 
#for that purpose what to use mean or median? 
#definitely median becoz mean can get affected by over-all skew
#now we have to find out median review rating within each category and then will replace that with each missing value
```

```python
#code for replacing that missing value
df["Review Rating"] = df.groupby("Category")["Review Rating"].transform(lambda x:x.fillna(x.median()))
```

```python
#checking if null values left
df.isnull().sum()
```

```python
#snakcasing->all lower case with underscores eg cutsomer_id , promo_code_used
#formatting column names in snake casing so that it will be easy to work with that names
#make code consistent and easy to use 
df.columns=df.columns.str.lower()  # making names in lower case first 
df.columns=df.columns.str.replace(" ", "_") # replacing space with underscore
```

```python
df.columns
```

```python
#still column name ->purchase_amount_(usd) is not looking nice so we will change it with different name 
df=df.rename(columns={"purchase_amount_(usd)":"purchase_amount"})
```

```python
df.columns
```

```python
#creating age_groups column
labels=["Young Adult","Adult","middle-aged","Senior"]
df["age_group"]=pd.qcut(df["age"],q=4,labels=labels)

```

```python
df[["age","age_group"]].head(10)
```

```python
df.head()
```

```python
#create purchase_frequency_days

frequency_mapping={
    "Fortnightly":14,
    "Weekly":7,
    "Monthly":30,
    "Quarterly":90,
    "Bi-Weekly":14,
    "Annually":365,
    "Every 3 Months":90
    
}
df["purchase_frequency_days"]=df["frequency_of_purchases"].map(frequency_mapping)
```

```python
df[["purchase_frequency_days","frequency_of_purchases"]].head(10)
```

```python
#is discount_applied and promo_code_used column same?
#lets check 
(df["discount_applied"]==df["promo_code_used"]).all()
```

```python
#got true for above check -->means both columns carry same information
#so lets remove one column
df=df.drop ("promo_code_used",axis=1)
```
